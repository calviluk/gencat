Lets start importing the rows from the db file

```python
from ase.db import connect
import os
```

Lets combine all data bases into one single data base

```python
db_paths = [
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures2.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures3.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures4.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures5.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures6.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures7.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures8.db',
    './extended_relaxed_structures/extended_relaxed_structures/extended_relax_structures9.db',
]

combined_path = './combined_relax_structures.db'
if os.path.exists(combined_path):
    os.remove(combined_path)

db_combined = connect(combined_path)

for path in db_paths:
    db = connect(path)
    for row in db.select():
        db_combined.write(row.toatoms(), key_value_pairs=row.key_value_pairs)

print(f"Combined database has {len(db_combined)} entries.")
```

The code below deletes the output file if it exists and prints the lenght of both input and output data base files. This serves just as a general check 

```python
# This are the keys that where printed into the db file in Main_SUN.py
# structures[i] 
# avg_energy_above_hull_per_atom=avg_energy_above_hull_per_atom
# avg_rmsd_from_relaxation=avg_rmsd_from_relaxation
# frac_novel_unique_stable_structures=frac_novel_unique_stable_structures
# frac_stable_structures=frac_stable_structures
# frac_successful_jobs=frac_successful_jobs
# avg_comp_validity=avg_comp_validity
# avg_structure_comp_validity=avg_structure_comp_validity
# avg_structure_validity=avg_structure_validity
# frac_novel_structures=frac_novel_structures
# frac_novel_systems=frac_novel_systems
# frac_novel_unique_structures=frac_novel_unique_structures
# frac_unique_structures=frac_unique_structures
# frac_unique_systems=frac_unique_systems
# precision=precision
# recall=recall 
# c_id=i

out_path = './S_filtered_structures.db'
if os.path.exists(out_path):
    os.remove(out_path)
db_in = connect('./combined_relax_structures.db')
db_out = connect(out_path)
print(f"Input database has {len(db_in)} entries.")
print(f"Output database has {len(db_out)} entries.")
```

The code below filters the structures that have an energy above hull below 0.1 eV and writes them into a data base 

```python
for row in db_in.select('avg_energy_above_hull_per_atom<0.1',frac_novel_structures=1.0,frac_unique_structures=1.0):
    atoms = row.toatoms()  # get the atomic structure
    # Copy all metadata (key-value pairs)
    db_out.write(atoms, key_value_pairs=row.key_value_pairs)
    # print(f"Wrote ID={row.id}, formula={row.formula}, energy={row.avg_energy_above_hull_per_atom:.4f}")
```

Finally this part of the code prints out the lenght and each of the rows in the output file

```python
for row in db_out.select():
    print(row.id, row.formula, row.avg_energy_above_hull_per_atom, row.frac_novel_structures, row.frac_unique_structures)

print(f"Output database has {len(db_out)} entries.")
```
